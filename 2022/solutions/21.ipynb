```python
import re
import operator
import time
import copy

ops = {
    '+' : operator.add,
    '-' : operator.sub,
    '*' : operator.mul,
    '/' : operator.truediv,  # use operator.div for Python 2
    '%' : operator.mod,
    '^' : operator.xor,
    '=' : operator.eq,
}

def get_diff(i, o, calcs):
    output = copy.deepcopy(o)
    output['humn'] = i
    while True:
        for name, action in calcs.items():
            l, op, r = action
            if l in output and r in output and name not in output:
                if name != 'root':
                    output[name] = float(ops[op](output[l], output[r]))
                else:
                    return output[l]-output[r]

                
with open('../inputs/21.txt') as f:
    o = {}
    calcs = {}
    for line in f:
        name, action = line.split(':')

        try:
            o[name.strip()] = int(action.strip())
        except:
            result = re.search(r"([a-z]+) (.) ([a-z]+)", action.strip())
            calcs[name.strip()] = result.groups()
            
    # mods
    # at least for my pinput, the right value stayed the same and the left
    # input decreased as numbers got larger. so the goal was using that info
    # to iteratively pick smaller and smaller windows, finding the value where
    # the left becomes smaller than the right, and then adjusting.
    calcs['root'] = (calcs['root'][0], '=', calcs['root'][2])

    count = 0
    

    # ub will be increased inside main loop til we find a range
    # containing the answer
    lb, ub = 0, 1000
    
    print(ub-lb)
    range_ =  (ub - lb) // 100
    answer = None
    diff = None
    while not answer:
        print('new params', lb, ub, range_, diff)
        
        overshot = False
        for i in range(lb, ub, range_):
            diff = get_diff(i, o, calcs)
            if diff == 0:
                answer = i
            if diff < 0:
                # when we go negative, we need to restart at the last positive value
                # for these params. we can also increase the granulariy since the 
                # range of numbers will be much smaller
                lb = i - range_
                range_ = range_ // 10
                overshot = True
                break
          
        # if we still haven't found the range, increase the ub until we
        # find a number that leads to a negative diff. once we have the
        # negative diff, then we can progressively narrow the window til
        # we find the answer.
        if not overshot:
            range_ *= 10
            ub += range_ * 10
        
    print('answer', answer)
```
