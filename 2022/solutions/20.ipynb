```python
# Link list Node Class
class Node:
    def __init__(self, data, id_):
        self.id = id_
        self.data = data
        self.prev = None
        self.next = None
 
 
class DoublyLL:
    def __init__(self):
        self.head = None
        self.tail = None
        self.id = 0
 
    # Function to print linked list
    def print_list(self):
        temp = self.head
        while temp:
            print(temp.data, end=" ")
            temp = temp.next
        print()
 
    # Function to push a node in DLL
    def push(self, item):
 
        # DLL is empty
        if self.tail == None:
            temp = Node(item, self.id)
            self.tail = temp
            self.head = temp
        # DLL is not empty
        else:
            temp = Node(item, self.id)
            self.tail.next = temp
            temp.prev = self.tail
            self.tail = temp
        self.id += 1

    def find_by_id(self, i):

        temp = self.head
 
        # Traversing the list
        while temp != None:
            if temp.id == i:
                return temp
            temp = temp.next
 
        return None

    def find_by_data(self, data):

        temp = self.head
 
        i = 0
        # Traversing the list
        while temp != None:
            if temp.data == data:
                print('found', i)
                return temp
            temp = temp.next
            i += 1
 
        return None

    def find_insertion_point(self, a):
        temp = a
        moves = a.data % (self.id - 1)


        om = moves
        while moves > 0:
            temp = temp.next
            if not temp:
                temp = self.head
            moves -= 1
#         print('ins', a.data, 'in', om, 'moves', temp.data)
        return temp
 
    # Function to swap the nodes
    # consisting of x and y
    def remove_and_insert_after(self, to_move, insert_after):
 
        # Edge Cases
        if self.head == None or self.head.next == None or to_move == insert_after:
            return

        if to_move == self.head:
#             print('new head')
            self.head = to_move.next
        if to_move == self.tail:
#             print('new tail')
            self.tail = to_move.prev
 
        # first remove item being moved and ensure
        # the 
        if to_move.next:
#             print('tomove next')
            to_move.next.prev = to_move.prev 
        if to_move.prev:
#             print('tomove prev')
            to_move.prev.next = to_move.next 
        
        # insert between 
        insert_before = insert_after.next
        
        to_move.next = insert_before
        to_move.prev = insert_after
        
        
        if insert_after:
            insert_after.next = to_move
        else:
            self.head = to_move
            
        if insert_before:
#             print('no bef')
            insert_before.prev = to_move
        else:
#             print('no aft')
            self.tail = to_move
         
def main():
    num = 811589153
    with open('../inputs/20.txt') as f:
        dll = DoublyLL()
        i = 0
        for line in f:
            dll.push(num * int(line.strip()))
            i+=1


        for i in range(10):
#             dll.print_list()
            print(i)
            seen = set()

            swaps = 0
            while len(seen) < dll.id:
                a = dll.find_by_id(len(seen))
                seen.add(a.id)

                if a.data == 0:
                    continue
                b = dll.find_insertion_point(a)
                dll.remove_and_insert_after(a, b)
                swaps += 1
        
        l = []
        tmp = dll.head
        while tmp:
            l.append(tmp.data)
            tmp = tmp.next
        
        targets = [1000, 2000, 3000]
        s = 0
        for t in targets:
            idx = l.index(0) +  (t % len(l))
            idx = idx % len(l)
            val = l[idx]
            s += l[idx]
            
        print('answer', s)


main()
```
