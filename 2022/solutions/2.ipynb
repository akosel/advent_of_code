```python
class Action:
    mapping = {
        'A': 'Rock',
        'B': 'Paper',
        'C': 'Scissors',
    }
    
    results = {
        'X': 'Lose',
        'Y': 'Draw',
        'Z': 'Win',
    }
    
    beats = {
        'Rock': 'Scissors',
        'Paper': 'Rock',
        'Scissors': 'Paper',
    }
    
    loses_to = { v: k for k, v in beats.items() }
    
    points = {
        'Rock': 1,
        'Paper': 2,
        'Scissors': 3
    }
    
    def __init__(self, action=None, result=None):
        mapped_action = self.mapping[action]
        if action and result:
            if self.results[result] == 'Lose':
                self.action = self.beats[mapped_action]
            elif self.results[result] == 'Win':
                self.action = self.loses_to[mapped_action]
            else:
                self.action = mapped_action
            
        elif action:
            self.action = mapped_action
        
    def __gt__(self, other):
        return self.beats[self.action] == other.action
    
    def __eq__(self, other):
        return self.action == other.action
        
    def __lt__(self, other):
        return self.beats[other.action] == self.action
    
    
    
    def get_points(self, other):
        base = self.points[self.action]
        if self > other:
            return 6 + base
        elif self == other:
            return 3 + base
        
        return 0 + base
    
    def __str__(self):
        return self.action
   

with open('../inputs/2.txt') as f:
    counter = 0
    for line in f:
        action, result = line.strip().split(' ')
        
        them, us = Action(action), Action(action, result)
        
        points = us.get_points(them)
        print(us, them, Action.results[result], points)
        
        counter += points
        
print(counter)
        
```
