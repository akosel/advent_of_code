```python
import re
import copy
import time
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
 
from collections import defaultdict

from IPython.display import display, HTML, clear_output
display(HTML("<style>.container { width:100% !important; }</style>"))


MAX_T = 31

def can_eventually_buy(cost, robots):
    for idx, robot in enumerate(robots):
        if cost[idx] and not robot:
            return False
        
    return True

def maximize(costs, state, max_):   
    t, resources, robots = state
    if t == MAX_T:
        return resources[3] + robots[3]
    
    # if number of robots is greater than the most it costs per turn.
    for idx, robot_count in enumerate(robots):
        if idx == 3:
            continue
            
        required = [cost[idx] for cost in costs]
        max_required = max(required)
    
        if robot_count > max_required:
            return 0
    
    states = []
    
    # no need to test buying other types of robots if we can buy the geode bot
    # once per turn
    
    backwards = reversed([(idx, cost) for idx, cost in enumerate(costs)])
    
    for idx, cost in backwards:
#         if idx < 3 and robots[3] > 1:
#             continue
        # stop making clay and ore robots after we have a certain number of
        # obsidian robots
        if idx < 2 and robots[2] > 3:
            continue

        if not can_eventually_buy(cost, robots):
            continue
                
        t2 = t
        new_resources = np.copy(resources)
        while not all(new_resources >= cost):
            t2 += 1
            new_resources += robots
            if t2 >= MAX_T:
                break
        
        if t2 >= MAX_T:
            state = (t2, new_resources, robots)
            if not states or not states[0][0] == t2:
                states.append(state)
            continue
            
        # simulate next turn where once we have enoguh to buy the given robot
        new_robots = robots.copy()
        new_resources += robots
        new_robots[idx] = new_robots[idx] + 1
        new_resources -= cost
        
        states.append((t2+1, new_resources, new_robots))
    
    
    for state in states:
        score = maximize(costs, state, max_)
        if score > max_[0]:
            max_[0] = score
            max_state = state
            print('new high score', max_, max_state)
            
    return max_[0]
    
    
    

with open('../inputs/19.txt') as f:
    total = 0
    i = 1
    scores = []
    for line in f:
        print(line)
        result = re.search(r"Each ore robot costs ([0-9]+) ore. Each clay robot costs ([0-9]+) ore. Each obsidian robot costs ([0-9]+) ore and ([0-9]+) clay. Each geode robot costs ([0-9]+) ore and ([0-9]+) obsidian.", line)
        groups = [int(i) for i in result.groups()]
        
        costs = np.array([
            np.array([groups[0], 0, 0, 0]),
            np.array([groups[1], 0, 0, 0]),
            np.array([groups[2], groups[3], 0, 0]),
            np.array([groups[4], 0, groups[5], 0]),
        ])


        state = (0, np.array([0, 0, 0, 0]), np.array([1,0,0,0]))
        score = maximize(costs, state, [0])
        
        print(i, score)
        scores.append(score)
        i += 1
        
        if i > 3:
            break
    print('answer', math.prod(scores))
    
    
    
```
